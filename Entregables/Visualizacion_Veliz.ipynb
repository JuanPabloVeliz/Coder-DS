### Importación de librerías, carga y descripción del dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/JuanPabloVeliz/Coder-DS/blob/main/Dataset/train.csv?raw=true")
df.describe().round(2)
```

### Gráficos con Matplotlib

```python
fig = plt.figure(figsize=(20, 5))
ax1 = fig.add_subplot(1,3,1)
ax2 = fig.add_subplot(1,3,2)
ax3 = fig.add_subplot(1,3,3)

ax1.hist(x=df["Loan Amount"], bins=[0,5000,10000,15000,20000,25000,30000,35000])
ax1.set_title("1.Cantidad de Préstamos por Importe")
ax1.set_xlabel("Importe en USD")
ax1.set_ylabel("Cantidad de Préstamos")

ax2.scatter(x=df["Revolving Balance"], y=df["Revolving Utilities"], alpha=0.5)
ax2.set_title("2.Relación Balance Recurrente / % de línea de crédito utilizada")
ax2.set_xlabel("Balance recurrente")
ax2.set_ylabel("% Línea de crédito utilizada")

ax3.plot(df["Loan Amount"].groupby(df["Grade"]).mean())
ax3.set_title("3.Promedio de Importe de Préstamos")
ax3.set_xlabel("Segmento")
ax3.set_ylabel("Importe en USD")
ax3.set_ylim(0,20000)
```

#### Interpretación
Gráfico 1. Cantidad de Préstamos por Importe

    Los datos contenidos en el dataset indican que agrupando los préstamos según su importe, los más frecuentes son los que corresponden al rango entre USD 15.000 y USD 20.000, y los menos frecuentes son los menores a USD 5.000 y los mayores de USD25.000

Gráfico 2. Relación Balance Recurrente / % de línea de crédito utilizada

    En el gráifco se observa que no existe una relación directa entre el balance de deuda recurrente y el ratio de utilización de la línea de crédito asignada al cliente

Gráfico 3. Promedio de Importe de Préstamos según Segmento Socioeconómico

    Analizando los importes promedios de los préstamos por segmento, no se advierten diferencias significativas entre ellos. Teniendo en cuenta también el gráfico anterior, puede interpretarse que los acreedores tienen mayor tendencia a ajustar el riesgo de "default" con la tasa de interés que con límites de montos de préstamos. Es decir que prefieren otorgar préstamos y asumir mayor el riesgo cobrando una mayor tasa, antes que limitar el otorgamiento de crédito para reducir su exposición al riesgo.

### Gráficos con Seaborn

```python

fig = plt.figure(figsize=(20, 5))
ax1 = fig.add_subplot(1,3,1)
sns.histplot(df["Interest Rate"])

ax2 = fig.add_subplot(1,3,2)
sns.boxplot(x=df["Grade"].sort_values(), y=df["Interest Rate"])

ax3 = fig.add_subplot(1,3,3)
sns.scatterplot(y=df["Interest Rate"], x=df["Loan Amount"], alpha=0.1)
```

#### Interpretación

Gráfico 1. Cantidad de Préstamos por Tasa de Interés

    Los datos contenidos en el dataset indican que agrupando los préstamos según su tasa de interés, lo más frecuente es que la tasa esté en el rango entre 11% y 12%.


Gráfico 2. Tasa de Interés Según Segmento Socioeconómico

    El gráfico muestra en concordancia con el anterior, que la tasa más común es aproximadamente de 11-12%, independientemente del segmento del deudor. También se aprecia que el límite inferior es del 5%, y el superior del 21%, con algunos casos aislados por encima de ese valor.


Gráfico 3. Relación Entre Tasa de Interés e Importe del Préstamo
    
    El gráfico muestra la relación entre los importes de los préstamos y las tasas de interés. La distribución indica que no existe una relación directa entre ambas variables.
