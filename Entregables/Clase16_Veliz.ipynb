```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/JJTorresDS/ds-data-sources/main/income_per_person_gdppercapita_ppp_inflation_adjusted.csv")
```

```python
df.head()
```

```python
df_ar = df.loc[df["country"] == "Argentina"]
df_ar = pd.melt(df_ar, id_vars=["country"], var_name="year", value_name='GDP')
df_ar["year"] = df_ar["year"].astype(int)
df_ar = df_ar.loc[df_ar["year"] > 2014].loc[df_ar["year"] < 2022]
df_ar
```

```python

df_ar["GDP"] = df_ar["GDP"].astype(str)

def value_to_float(x):
    if "k" in x:
        return float(x.replace("k", "")) * 1000
    return float(x)
    
df_ar["GDP"] = df_ar["GDP"].apply(value_to_float)
```

```python
avg_GDP = df_ar["GDP"].mean()
print("GDP promedio Argentina aÃ±os 2015-2021: ", avg_GDP)
```
